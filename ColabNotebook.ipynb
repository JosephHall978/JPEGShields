```python
from google.colab import drive
import os

drive.mount('/content/drive')
shared_folder_path = '/content/drive/My Drive/Watermark Evaluation Dataset-Public'
sources = os.listdir(shared_folder_path)
```

#Example

```python
import matplotlib.pyplot as plt
import cv2 as cv

source = sources[0]
source_path = os.path.join(shared_folder_path, source)
image = os.path.join(source_path, os.listdir(source_path)[0])
img = cv.imread(image)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
from JPEGShields.WaterMarker import WaterMarker
import numpy as np

marker = WaterMarker()
watermark = np.random.randint(0, 2, size=100, dtype=np.uint8)
img = marker.generate(img,watermark)
plt.imshow(img)
```
