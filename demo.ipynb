```python
import matplotlib.pyplot as plt
from skimage import data
import numpy as np
from WaterMarker import WaterMarker

wm = WaterMarker()
original_img = data.astronaut()

np.random.seed(42)
test_watermark = np.random.randint(0, 2, 100)

watermarked_img = wm.generate(original_img, test_watermark)

bits_lost = wm.evaluate(watermarked_img, test_watermark)
metrics = wm.evaluate_watermarking(original_img, watermarked_img)
print(f"Evaluation: {bits_lost} bits mismatched out of 100.")
print(f"{metrics}")

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(original_img)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(watermarked_img)
axes[1].set_title("Watermarked Image (DFT Diagonal)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
